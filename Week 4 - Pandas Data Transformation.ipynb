```python
dff=df.dropna()
```

```python
dff.sort_values("Calories", ascending=False)
```

```python
dff
```

```python
dff.reset_index()
```

```python
agg_dff = dff.groupby("Duration").agg(
    min_pulse_max=('Pulse','max'),
    Average_Calories=('Calories','mean')
).reset_index()
```

```python
print("Aggregated DataFrame")
agg_dff
```

```python
agg_dff2 = dff.groupby("Duration").agg(
    Max_pulse_max=("Maxpulse", "max")
).reset_index()
agg_dff2
```

```python
merged_dffs=pd.merge(agg_dff,agg_dff2,on='Duration',how='left')
```

```python
merged_dffs
```

```python
concat_dfs=pd.concat([merged_dffs,agg_dff,agg_dff2], ignore_index=False)
```
