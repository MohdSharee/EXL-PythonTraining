```python
import pandas as pd
df=pd.read_csv("data.csv")
df.head()
```

```python
df1=df.sort_values("Duration")
df1
df2=df1.dropna()
df2.reset_index()
df3=df2[:28]
df3
```

```python
plt.bar(df['Pulse'],df['Maxpulse'])
plt.show()
```

```python
plt.plot(df3['Duration'],df3['Calories'],"ro")
plt.xlabel("Duration of exercise")
plt.ylabel("Calories burnt")
plt.axis((0,70,150,520))
plt.show()
```

```python
import numpy as np
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
