```python
import seaborn as sns

sns.set_theme()

tips=sns.load_dataset("tips")

sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
import pandas as pd
df=pd.read_csv("data.csv")
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.set_theme()
sns.lineplot(
    data=df,
    x="Duration", y="Date"
)
```

```python
sns.relplot(
    data=df, x="Pulse", y="Calories",
    hue="Pulse"
)
```

```python
sns.boxplot(
    data=df, x="Calories"
)
```

```python
sns.barplot(
    data=df, x="Pulse", y="Date",
    hue="Pulse"
)
```

```python

sns.pairplot(
    data=df, hue="Duration", palette="coolwarm"
)
```

```python
df_pivot=df.pivot(columns="Pulse", values="Calories")
sns.heatmap(df_pivot, annot=True, fmt="d", cmap="YlGnBu")
```
